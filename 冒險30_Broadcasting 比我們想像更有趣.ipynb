1. 現在我們想要同時算很多人的BMI 值，假設有兩個班級各有三位同學，每位同學的身高體重你可以隨便想，或是用更高級的亂數取（要在合理範圍），接著套用BMI公式就完成！如果你覺得這樣太簡單，可以試試看找出兩班同學的平均BMI 值。

```
import numpy as np
weight = np.array([[40,75,68], # 第一班同學體重
                   [50,66,83]]) # 第二班同學體重
height = np.array([[156,183,175], # 第一班同學身高
                   [165,178,192]]) # 第二班同學身高

bmi = weight/(height/100)**2

mean = bmi.mean(axis=1)
mean
```

2. 我們都知道怎麼判斷一個數是2 的倍數、3 的倍數或5 的倍數，但是你知道怎麼判斷7 的倍數嗎？事實上，只要把一個數字從右到左分別乘以 1, 3, 2, 6, 4, 5 （有需要則重複）之後再加起來，如果得到的數字可以被7 整除，那麼此數就是7 的倍數喔！現在請你隨便想一個數字，越複雜越好，這才看得出array 的偉大之處，接著用上面提過的「內積」方法來測試它是不是7 的倍數。

```
num = '516469486' # 若換成input的形式?
arr = np.array([int(i) for i in num])
w = np.array([2,3,1,5,4,6,2,3,1]) # 如何更自動化的寫出w?
if arr @ w % 7 == 0:
  print(f'{num} is the multiple of 7')
else:
  print(f'{num} is not the multiple of 7')
```
