1. 設計一個 class，是準備讓你記錄自己有興趣的物品、價格、商店。試著輸入一些資料，你可以想到什麼樣的應用情境嗎？

```
class Buy: # 此處內容可以自己修改
  COMM = ['台北東京', '台北宿霧', '台北夏威夷']
  PRICE = ['5950', '6240', '11950'] 
  SHOP = ['樂桃航空', '亞洲航空', '中華航空']
  def __init__(self, c, p, s):
    self.commodity = c
    self.price = p
    self.shop = s
  def __repr__(self):
    return self.SHOP[self.shop] + '-' + self.COMM[self.commodity] + '-' + self.PRICE[self.price]
```

```
ticket01 = Buy(0, 0, 0)
ticket01
```
