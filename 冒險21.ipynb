### 1. 設定一個串列，裡面的元素是 1、11、111、1111、11111、⋯⋯ 還記得 Python 有「無限長度」的整數嗎？所以你可以打到累了不想打為止。再來想辦法算出另一個串列，裡面的元素就是把前面的數字都平方，看看有沒有發現什麼有趣的事！

```
list1 = [int('1'*i) for i in range(1,10)] # 先決定第一個串列元素
list2 = list(map(lambda x: x**2, list1)) # 對第一個串列每個元素都取平方
list2
```

### 2. 你有五個想要買的東西，價格分別是1350、2400、800、7980、5000。今天發現發票中了 2000，請想辦法找出小於等於 2000 的價格。你馬上會發現，當想買的東西更多的時候，只篩選出價格不太方便。你有可能設計出可以篩出價格，又能知道是哪個物品的方式嗎？

```
price = [1350, 2400, 800, 7980, 5000] # 價格串列
low_price = list(filter(lambda x: x<=2000, price)) # 找出小於等於2000的價格
low_price
```

```
dict1 = {1350:'A', 2400:'B', 800:'C', 7980:'D', 5000:'E'} # 列出對應於價格的商品A,B,C,D,E
buy = [dict1[i] for i in list(filter(lambda x: x<=2000, dict1.keys()))] # 找出價格小於等於2000的商品
buy
```
