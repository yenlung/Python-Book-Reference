### 1. 用for迴圈畫出一棵聖誕樹

```
for i in range(5):
  print(' '*int((9-(2*i+1))/2),'*'*(2*i+1),' '*int((9-(2*i+1))/2), end='\n')
```

### 2. 你和朋友總共五個人一起去吃飯，每個人點的餐價格是：price = [100, 150, 120, 90, 145]，你們最後決定大家平分就好，所以請用for迴圈把總金額算出來，再求出平均數。

```
price = [100, 150, 120, 90, 145]
num = 0
for i in price:
  num+=i

mean = num/len(price)
int(mean)
```
