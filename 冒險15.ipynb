### 1. 改寫我們的拍拍機器人，當使用者說“Bye”的時候，不要再顯示「拍拍」，而是回應「再見！」或「Bye ！祝你一切順利！」等等。

```
message = ''
while message != 'Bye':
  message = input('> ')
  if '悲傷' in message:
    print("ヾ(·∀·‵): 拍拍")
  elif '很快樂' in message:
    print("ヾ(·∀·‵): 太棒了")
  elif message == 'Bye':
    print("ヾ(·∀·‵): Bye ！祝你一切順利！")
  else:
    print("ヾ(·∀·‵): 嗯嗯")
```

### 2. 寫一個和電腦玩「剪刀、石頭、布」的遊戲，你可以用1 代表剪刀、2 代表石頭、3代表布，電腦也隨機選一個出來，然後判斷誰贏了。一直玩到使用者打入bye，才結束遊戲。

```
import numpy as np
player = ''
my_dict = {'scissors':1,'rock':2,'paper':3}

while player != 'Bye':
  player = input('> ')
  if player!='Bye':
    player = my_dict[player]
    n = np.random.choice(list(my_dict.keys()),1)[0]
    computer = my_dict[n]
    print(computer)
    if player==1 and computer==3:
      print("ヾ(·∀·‵): 玩家勝")
    elif player==3 and computer==1:
      print("ヾ(·∀·‵): 電腦勝")
    elif player < computer:
      print("ヾ(·∀·‵): 電腦勝")
    elif computer < player:
      print("ヾ(·∀·‵): 玩家勝")
    else:
      print("ヾ(·∀·‵): 平手")
```
