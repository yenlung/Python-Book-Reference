### 1. 我們的拍拍機器人，改成如果輸入的句子中有「悲傷」，才回應「拍拍」；如果有「很快樂」就回應「太棒了」；其他句子就回覆「嗯嗯」。當然你可以想辦法分更多的情況！

```
message = input("> ")
if '悲傷' in message:
  print("ヾ(·∀·‵): 拍拍")
elif '很快樂' in message:
  print("ヾ(·∀·‵): 太棒了")
else:
  print("ヾ(·∀·‵): 嗯嗯")
```

### 2. 讓使用者輸入身高、體重。幫他算出BMI 值，並且依BMI 值來告訴他是正常、過重或過輕。

```
message = input("請輸入身高體重: ")
height = int(message.split(',')[0])
weight = int(message.split(',')[1])
bmi = weight/(height/100)**2
if 18.5<= bmi or bmi<=24: 
  print("BMI檢測器: 正常")
elif bmi<18.5:
  print("BMI檢測器: 過輕")
else:
  print("BMI檢測器: 過重")
```
