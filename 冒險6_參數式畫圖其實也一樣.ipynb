在參數式畫圖的這段程式碼：

r = 1 - np.sin(θ )

x = r * np.cos(θ )

y = r * np.sin(θ )

我們可以發現重點就是「設計半徑 r 這個函數」。想辦法改變一下 r 的定義，看能不能更有意識的去設計出一些有趣的圖來。

```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
π = np.pi
θ = np.linspace(0, 2*π, 500)
r = 1 - np.sin(θ)*np.cos(θ)
x = r * np.cos(θ)
y = r * np.sin(θ)

ax = plt.gca()
ax.set_aspect('equal')
plt.plot(x, y)
```
