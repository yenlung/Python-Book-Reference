### 1. 寫一個函式，有兩個引數，一是身高、一是體重。回傳這個身高體重下的BMI 值。

```
def BMI(height, weight):
  bmi = weight/(height/100)**2
  return bmi
```

### 2. 寫一個函式，判斷一個數是不是質數。

```
def prime(num):
  ans=''
  if type(num) == int and num >1:
    for i in range(2,num):
      if num % i == 0:
        ans='not prime'
  else:
    ans='wrong number'  

  if ans=='':
    ans='prime'
  return ans
```

### 5. 我們的身份證是由一個英文字母，外加9 個數字組成。而第9 個數字其實是檢查碼，也就是說英文字母和前8 個數字，會決定第9 個數字是什麼。請上網查一下這規則是什麼，寫一個Python 函式，讓使用者輸入自己的身份證字號後，可以核對這是不是正確的身份證字號。

```
def identifier(id): # 請輸入字串形式的身分證字號
  d=0
  word = [chr(i) for i in range(65,91)]
  num = ['10','11','12','13','14','15','16','17','34','18','19','20','21',\
         '22','35','23','24','25','26','27','28','29','32','30','31','33']
  my_dict={}
  for i in range(len(word)):
    my_dict[word[i]] = num[i]

  first = my_dict[id[0]]
  d+=int(first[0])*1

  d+=int(first[1])*9

  for i in range(len(id[1:])):
    d+= int(id[1:][i])*(8-i)
  if d%10==0:
    last = 0
  else:
    last = 10-d%10
  return last
```

### 6. 寫一個函式，執行了就會開始猜數字遊戲。由電腦想一個1 到100 的數字，由使用者來猜。使用者沒猜到的話，電腦要告訴他太大或是太小，然後使用者再猜，一直到猜到為止。

```
import numpy as np

def guessNumber():
  ans = np.random.randint(1,101)
  message=''
  while message!=ans:
    message=int(input('請輸入一個1到100的數字: '))
    if ans<message:
      print("數字太大!")
    elif message<ans:
      print("數字太小!")
    else:
      print("Bingo!")
```
