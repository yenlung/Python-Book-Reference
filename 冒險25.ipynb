### 2. 我們來做一個幫助你背單字的程式。先用字典檔把你想記的英文單字（當然日文、韓文或其他任何語言也都可以）蒐集起來，每次隨機顯示一個單字（當然最好顯示大大的），然後用input() 等待，到使用者按enter 後才顯示中文翻譯。

```
import numpy as np
from IPython.display import Markdown

word_dict = {'resurgence':'復興',
             'pandemic':'大流行病',
             'hassle':'麻煩'}

word = np.random.choice(list(word_dict.keys()),1)[0]
obj = f'## {word}'
display(Markdown(obj))
useless = input()
ans = f'## {word_dict[word]}'
display(Markdown(ans))
```
