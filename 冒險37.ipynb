### 1. 找一些你感興趣的事情用array 記錄下來，再用DataFrame 呈現。舉例來說，我想知道近20 年來各大電影獎項最佳影片的主題，像是奧斯卡可能有1 部喜劇片、2部科幻片等等（這是亂掰的），來分析電影風格的轉變。或是蒐集2019-2020 這一季NBA 各隊對戰成績，來預測今年NBA 總冠軍等等。

```
# 現在我們有能力完成這題的前半部分(用DataFrame記錄感興趣的事)，後半部分等到第3篇的時候再來做
```

```
import numpy as np
import pandas as pd

jabbar = np.array([6,6,2,10,19])
jordan = np.array([6,5,6,10,14])
james = np.array([3,4,3,12,16])
magic = np.array([5,3,3,9,12])
duncan = np.array([5,2,3,15,15])

arr_df=[]
player = [jabbar, jordan, james, magic, duncan]
for i in player:
  arr_df.append(pd.DataFrame(i).T)
df = pd.concat(arr_df, axis=0)
df.columns = ['總冠軍次數', 'MVP次數', 'FMVP次數', 'NBA第一陣容次數', '全明星賽次數']
df.index = ['jabbar','jordan','james','magic','duncan']
df
```
