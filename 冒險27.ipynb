### 1. 還記得會寫詩的文青機器人嗎？我們想要把這作品存下來，所以請修改一下程式，讓 Python 會存下我們文青機器人的大作。當然，有可能你不會想要蓋掉先前的作品，所以在開檔時，可以用 'a' 這個參數，而不是用 'w'。'a' 是 append 的意思，這樣如果沒有這個檔名的話，會開個新檔，有這個檔，不會把原檔蓋過去，而是把新寫入的再加上去。

```
rawdata = '''我
我的
眼睛
妳
妳的
心
溫柔
日子
雨
風
天空
雲
等待
哭泣
戀愛
相遇
分離
忘記
心醉
驀然
吹過
思念
靈魂
停止'''

words = rawdata.split('\n')
```

```
import numpy as np

def poem():
  n = np.random.randint(2, 8)
  with open('poem.csv', 'a') as f:
    for i in range(n):
      m = np.random.randint(1, 6)
      sentence = np.random.choice(words, m, replace=False)
      print(" ".join(sentence), file=f)
```

```
poem()
```
