```
import matplotlib.pyplot as plt 
from sklearn.datasets import load_boston
```

把波士頓房價的資料拿進來

```
boston = load_boston()
boston.feature_names
```

看一下資料說明

```
print(boston.DESCR)
```

挑九個我們喜歡的 features 出來畫圖，用 subplot 確保他們分開來畫

```
boston_select = ['CRIM', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'PTRATIO', 'LSTAT']
X = boston.data
Y = boston.target

plt.figure(figsize=(8, 10))
for i, feature in enumerate(boston_select):
    plt.subplot(5, 3, i+1)
    plt.scatter(X[:, i], Y, s=1)
    plt.ylabel('price')
    plt.xlabel(feature)
    plt.tight_layout()
```
