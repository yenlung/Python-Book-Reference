### 2. 先用變數紀錄使用者的年齡、身高、體重，計算他的 BMI 值。假設一位 20 歲使用者算出的 BMI 值為 20.6，則輸出： " 你今年 20 歲，BMI 值為 20.6。"

```
age = 21
height = 165
weight = 53
bmi = weight/(height/100)**2
# round(bmi,1)
print('你今年',age,'歲','BMI值為',bmi)
```
