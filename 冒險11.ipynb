### 2. 當然，input 也可以寫更「有用」的程式。比方說，我們可以寫讓使用者輸入身高、體重，就幫他算出BMI 值的程式。

```
message = input("請輸入身高體重: ")
height = int(message.split(',')[0])
weight = int(message.split(',')[1])
print("BMI: ", weight/(height/100)**2)
```
