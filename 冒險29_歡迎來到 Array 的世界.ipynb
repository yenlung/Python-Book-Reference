1. 炭治郎和彌豆子最近想去冰島看極光，他們決定每晚住宿的預算是2500到3000台幣，現在他們在訂房網看中四間旅館，每晚價格分別是[11210, 14816, 10751, 8228]克朗，刷信用卡還能額外享有台幣1%現金回饋，我們已知1克朗 = 0.23655台幣，請用array來計算哪間旅館符合他們的預算。

```
import numpy as np

price = np.array([11210,14816,10751,8228])
price = price*0.23655*0.99
result = (2500<=price) & (price <=3000)
result
```
