```
import numpy as np

arr = np.zeros((8,8)).astype(int)
even = [i for i in range(8) if i%2==0]
odd = [i for i in range(8) if i%2!=0]
mask = [(i,j) for i in even for j in even]
mask.extend([(i,j) for i in odd for j in odd])
for i in mask: 
  arr[i] = 1
```

```
for row in arr:
  print()
  for idx in row:
    if idx==1:
      print('■',end=' ')
    else:
      print('□',end=' ')
```
