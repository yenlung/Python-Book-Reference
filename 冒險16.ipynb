### 由於這題題目過長，因此不放上來

```
%matplotlib inline
import matplotlib.pyplot as plt

message = int(input('請輸入一個正整數: '))
box = [message]
while message != 1:  
  if message%2==0:
    message = int(message/2)
  else:
    message = 3*message+1
  box.append(message)

plt.plot(box)
```
